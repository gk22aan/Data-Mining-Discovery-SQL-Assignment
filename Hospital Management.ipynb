```python
# Generate Patient Details Table
import csv
from faker import Faker
import random
from gender_guesser.detector import Detector

# Function to guess gender based on the first name
def guess_gender_from_first_name(first_name):
    gender = gender_detector.get_gender(first_name)
    # Clean up the gender result
    if gender in ['mostly_male', 'mostly_female']:
        gender = gender.split('_')[1]
    return gender


def generate_patient_id(used_ids):
    new_id = f'1' + ''.join(np.random.choice(list(string.digits), 5))
    while new_id in used_ids:
        new_id = f'1' + ''.join(np.random.choice(list(string.digits), 5))
    used_ids.add(new_id)
    return new_id

fake = Faker()
gender_detector = Detector()

# Function to generate a list of medical history terms
def generate_medical_history(n):
    medical_history_terms = [
        "Hypertension",
        "Diabetes",
        "Asthma",
        "Allergies",
        "Heart Disease",
        "Arthritis",
        "Cancer",
        "Depression",
        # Add more medical history terms as needed
    ]

    # Generate a random medical history for each patient
    medical_history_list = [random.choice(medical_history_terms) for _ in range(n)]

    return medical_history_list

# Specify the number of patients
n = 200

# Generate fake patient data including medical history
patients_data = []
used_patient_ids = set()
for _ in range(n):
    first_name = fake.first_name()
    gender = guess_gender_from_first_name(first_name)
    patient_data = {
        "PatientID": generate_patient_id(used_patient_ids),
        "FirstName": first_name,
        "LastName": fake.last_name(),
        "DateOfBirth": fake.date_of_birth(minimum_age=18, maximum_age=90),
        "Gender": gender,
        "ContactNumber": (f'+1-' + ''.join(np.random.choice(list(string.digits),10))),
        "Address": fake.address(),
        "MedicalHistory": generate_medical_history(3),  # Generating 3 medical history terms per patient
        "DateAdmitted": fake.date_this_century(),
    }
    patients_data.append(patient_data)


df_patients_data = pd.DataFrame(patients_data)
df_patients_data.to_csv('PatientsDetail.csv',index=False)
```

```python
# generate Doctors Detail Table

import csv
from faker import Faker
import numpy as np
import string

fake = Faker()

# Function to generate unique IDs for Doctors
def generate_doctor_id(used_ids):
    new_id = f'1{"".join(np.random.choice(list(string.digits), 4))}'
    while new_id in used_ids:
        new_id = f'1{"".join(np.random.choice(list(string.digits), 4))}'
    used_ids.add(new_id)
    return new_id

# Function to generate email address for Doctors
def generate_random_email(name):
    domain = ['gmail.com', 'ac.co.uk', 'yahoomail.com', 'rediffmail.com']
    domains = np.random.choice(domain)
    return f"{name.replace(' ', '.').lower()}@{domains}"



# Specify the number of doctors
n_doctors = 200

# Generate fake doctor data
doctors_data = []
used_doctor_ids = set()
for _ in range(n_doctors):
    first_name = fake.first_name()
    last_name = fake.last_name()
    
    doctor_data = {
        "DoctorID": generate_doctor_id(used_doctor_ids),
        "FirstName": first_name,
        "LastName": last_name,
        "Specialization": fake.random_element(["Cardiology", "Orthopedics", "Pediatrics"]),
        "ContactNumber": (f'+1-' + ''.join(np.random.choice(list(string.digits),10))),
        "Email": generate_random_email(first_name),
        "DepartmentID": f'DepartmentID_{np.random.randint(1, 6)}',  # Assuming 5 departments for example
    }
    doctors_data.append(doctor_data)
# Export to CSV
df_doctors_data = pd.DataFrame(doctors_data)
df_doctors_data.to_csv('DoctorsDetail.csv',index=False)
```

```python
# code for prescriptions Table
import csv
from faker import Faker
import numpy as np
import string

fake = Faker()

# Function to generate unique IDs for Prescriptions
def generate_prescription_id(used_ids):
    new_id = f'1{"".join(np.random.choice(list(string.digits), 4))}'
    while new_id in used_ids:
        new_id = f'1{"".join(np.random.choice(list(string.digits), 4))}'
    used_ids.add(new_id)
    return new_id

# Function to generate relevant prescription notes
def generate_prescription_notes():
    # Customize the function based on specific medical conditions or advice
    medical_conditions = fake.random_elements(elements=("Hypertension", "Diabetes", "Asthma", "Arthritis"), length=2)
    advice = fake.random_element(elements=("Take medication before meals.", "Avoid strenuous activities.", "Get plenty of rest."))
    additional_info = fake.random_element(elements=("Follow up in two weeks.", "Stay hydrated.", "Call for any concerns."))

    return f"Medical conditions: {', '.join(medical_conditions)}. {advice}. {additional_info}"

# Specify the number of prescriptions
n_prescriptions = 200


# Generate fake prescription data using existing patient and doctor IDs
prescriptions_data = []
used_prescription_ids = set()
for _ in range(n_prescriptions):
    prescription_data = {
        "PrescriptionID": generate_prescription_id(used_prescription_ids),
        "DoctorID": np.random.choice(list(used_doctor_ids)),
        "PatientID": np.random.choice(list(used_patient_ids)),
        "DatePrescribed": fake.date_this_century(),
        "Notes": generate_prescription_notes(),
    }
    prescriptions_data.append(prescription_data)

# Export to CSV
df_prescriptions_data = pd.DataFrame(prescriptions_data)
df_prescriptions_data.to_csv('PrescriptionsDetail.csv',index=False)
```

```python
# code for Medication Table

import csv
from faker import Faker
import numpy as np
import string

fake = Faker()

# Function to generate unique IDs for Medications
def generate_medication_id(used_ids):
    new_id = f'M{"".join(np.random.choice(list(string.digits), 4))}'
    while new_id in used_ids:
        new_id = f'M{"".join(np.random.choice(list(string.digits), 4))}'
    used_ids.add(new_id)
    return new_id

# Function to generate relevant medication names
def generate_medication_name():
    return fake.random_element(elements=("Aspirin", "Ibuprofen", "Lisinopril", "Simvastatin", "Metformin"))

# Specify the number of medications
n_medications = 200

# Generate fake medication data
medications_data = []
used_medication_ids = set()
for _ in range(n_medications):
    medication_data = {
        "MedicationID": generate_medication_id(used_medication_ids),
        "MedicationName": generate_medication_name(),
        "Dosage": fake.random_element(elements=("5mg", "10mg", "20mg")),
        "Units": fake.random_element(elements=("Tablet", "Capsule", "Injection")),
    }
    medications_data.append(medication_data)
# Export to CSV
df_medications_data = pd.DataFrame(medications_data)
df_medications_data.to_csv('MedicationsDetail.csv',index=False)
```

```python

# Function to generate unique IDs for Departments
def generate_department_id(used_ids):
    new_id = f'DepartmentID_{len(used_ids) + 1}'
    used_ids.add(new_id)
    return new_id

# Specify department names
department_names = ["Cardiology", "Orthopedics", "Pediatrics", "Oncology", "Neurology"]

# Generate fake department data
departments_data = []
used_department_ids = set()

for department_name in department_names:
    department_data = {
        "DepartmentID": generate_department_id(used_department_ids),
        "DepartmentName": department_name,
    }
    departments_data.append(department_data)

# Export to CSV
df_departments_data = pd.DataFrame(departments_data)
df_departments_data.to_csv('DepartmentsMapping.csv', index=False)
```
